## Step 1: Data Acquisition

```python
# Importing the required Libraries
import pandas as pd
```

```python
filePath = 'https://archive.ics.uci.edu/ml/machine-learning-databases/autos/imports-85.data'
dfAuto = pd.read_csv(filePath, header = None)
```

```python
# Printing the first 5 rows in the Data Frame
dfAuto.head()
```

```python
# Check the column names
dfAuto.columns
```

```python
# Fetching the correct names using file "https://archive.ics.uci.edu/ml/machine-learning-databases/autos/imports-85.names"
headers = ['symboling','normalized-losses','make','fuel-type','aspiration','num-of-doors','body-style',\
           'drive-wheels','engine-location','wheel-base','length','width','height','curb-weight','engine-type', \
           'num-of-cylinders','engine-size','fuel-system','bore','stroke','compression-ratio','horsepower', \
           'peak-rpm','city-mpg','highway-mpg','price']
# Assigning Column names to the Data Frame
dfAuto.columns = headers
dfAuto.columns
```

```python
# Printing the first 5 rows of the Data Frame
dfAuto.head()
```

```python
# Replacing ? with NaN
import numpy as np
dfAuto.replace('?', np.nan, inplace = True)
dfAuto.head()
```

```python
# Saving the CSV to Local
dfAuto.to_csv('auto85.csv', index = False)
```

```python
# checking if there are any numm values in the Target Variable 'price'
dfAuto['price'].isnull().value_counts().to_frame()
```

```python
# Droping rows where price is NaN
dfAuto.dropna(subset=['price'], axis = 0, inplace = True)
```

```python
# Confirming if the values are dropped
dfAuto['price'].isnull().value_counts().to_frame()
```
